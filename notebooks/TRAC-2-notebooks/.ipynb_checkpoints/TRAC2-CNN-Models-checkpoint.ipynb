## Helper functions

```python
def from_prob_to_labels(model, x, task):
    '''
    Returns labels based on predicted probability on labels [CAG,NAG,OAG] for task A. Task B is binary, and 'GEN' represents 
    the positive class.
    Parameters:
    model: trained model
    x: input data
    task: either 'A' or 'B'
    '''
    pred = model.predict(x)
    
    index_a = {0:'CAG', 1:'NAG', 2:'OAG'}
    
    if task == 'A':
        highest_prob_class = np.argmax(pred, axis=1)
        labels = np.vectorize(index_a.get)(highest_prob_class.astype(int))
        
    elif task == 'B':
        labels = np.where(pred <0.5, 'NGEN', 'GEN')
    else:
        labels = []
        
    return labels    
```

```python
def to_binary_labels(string_labels, classes_list):
    '''
    Returns an array with 0 and 1 for a binary classification problem.
    Parameters:
    string_labels: array with 2 categories defined as strings e.g. ['cat', 'dog', 'dog', ...] 
    classes_list: array with the two classes. The order of the array defines which gets 0 and which gets 1. The first
                  gets 0.
    '''

    labels = label_binarize(string_labels, classes = classes_list).flatten()
    
    return labels 
```

```python
def to_one_hot_labels(string_labels):
    '''
    Returns one-hot encoded labels from a multi-class label vector e.g. ['cat', 'dog', 'dog', 'lion', 'cat', ...] 
    Parameters:
    string_labels: 
    '''
    labels = pd.get_dummies(string_labels)
    labels = labels.to_numpy()
    
    return labels
```

```python
def confusion_matrix_plot(model, input_data, true_labels, task, normalize=None):
    '''
    Returns a confusion matrix with a nice format.
    Parameters:
    model: trained model
    input data: data we want to use to evaluate the model
    true_labels: true labels 
    task: 'A' or 'B'
    normalize: if want to normalize the confusion matrix normalize='true'
    '''
    
    # get predicted labels
    pred_labels = from_prob_to_labels(model, input_data, task)
    
    # Create a confusion matrix
    cm = metrics.confusion_matrix(true_labels, pred_labels, normalize=normalize)
    cm = np.around(cm, 2)

    # Plot the confusion matrix
    if task == 'A':
        axis_labels = ['CAG', 'NAG', 'OAG']
    elif task == 'B':
        axis_labels = ['GEN', 'NGEN']

    fig, ax = plt.subplots(figsize=(4,4))
    im = ax.imshow(cm, cmap="Blues")

    # Create the ticks and labels
    ax.set_xticks(np.arange(len(axis_labels)))
    ax.set_yticks(np.arange(len(axis_labels)))
    ax.set_xticklabels(axis_labels)
    ax.set_yticklabels(axis_labels)

    # Axis titles
    plt.ylabel('True label', size=12)
    plt.xlabel('Predicted label', size=12)

    # Loop over data dimensions and create text annotations.
    for i in range(len(axis_labels)):
        for j in range(len(axis_labels)):
            text = ax.text(j, i, cm[i, j],ha="center", va="center", color="dimgrey", size=12)
    
    ax.set_title("Confusion Matrix", size=16, weight="bold")
    fig.tight_layout()
    plt.show()
```

```python
def loss_accuracy_plots(training_history, xrange, task):
    '''
    Returns plots for loss and accuracy during the training process of a NN.
    Parameters:
    training_history: object that stores the training history of the NN (from model.fit(...))
    xrange: range in x axis
    task: string used for the title in the plot
    '''
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))
    
    # loss plot
    ax1.plot(training_history.history['loss'], color='black')
    ax1.plot(training_history.history['val_loss'], color='blue')
    ax1.set_title('Training and validation loss Sub-Task ' + task)
    ax1.legend(['training', 'development'])
    ax1.grid(which='both')
    ax1.set_xticks(np.arange(0, xrange, 2))
    
    # accuracy plot
    if task == 'A':
        ax2.plot(training_history.history['categorical_accuracy'], color='black')
        ax2.plot(training_history.history['val_categorical_accuracy'], color='blue')
        ax2.set_title('Training and validation acccuracy Sub_Task ' + task)
        ax2.legend(['training', 'development'])
        ax2.grid(which='both')
        ax2.set_xticks(np.arange(0, xrange, 2))
    elif task == 'B':
        ax2.plot(training_history.history['binary_accuracy'], color='black')
        ax2.plot(training_history.history['val_binary_accuracy'], color='blue')
        ax2.set_title('Training and validation acccuracy Sub_Task ' + task)
        ax2.legend(['training', 'development'])
        ax2.grid(which='both')
        ax2.set_xticks(np.arange(0, xrange, 2))
    plt.show()
    
```

```python
# path to glove file- will use the embeddings with dimension = 300
glove_file ="../../../data/glove.6B.300d.txt"
```

```python
# create a vocabulary index 
# consider this maximum number of words- Played with larger vocab sizes, but 10,000 is enough.
max_tokens = 10000
# truncate or pad sequences to be this long
max_sequence_lenght = 100

vectorizer = TextVectorization(max_tokens=max_tokens, output_sequence_length=max_sequence_lenght)
```

```python
# train vocabulary
vectorizer.adapt(train_text)

# save vocabulary in a variable
vocab = vectorizer.get_vocabulary()
```

```python
print(f'Vocabulary has {len(vocab)} words.')
```

```python
# create an index for each word {word: index}
word_idx = dict(zip(vocab, range(len(vocab))))
```

```python
# Map words with their vector representation (embeddings)
embeddings_glove = {}
with open(glove_file) as f:
    for line in f:
        word, coefs = line.split(maxsplit=1)
        coefs = np.fromstring(coefs, "f", sep=" ")
        embeddings_glove[word] = coefs

print(f'Number of embeddings available: {len(embeddings_glove)}')
```

```python
# just to check let's get the dimensions of one of the embeddings
embeddings_glove['home'].shape
```

```python
# build embedding matrix to use it in the model
dimensions_emb = 300
# the plus two is for padding and unknown tokens
total_tokens = len(vocab) + 2
with_embedding = []
without_embedding = []

# initialize embedding matrix with zeroes
embedding_matrix = np.zeros((total_tokens, dimensions_emb))

for word, index in word_idx.items():
    emb_vector = embeddings_glove.get(word)
    # add to matrix
    # count converted and not converted words
    if emb_vector is not None:
        embedding_matrix[index] = emb_vector
        with_embedding.append(word)
    else:
        without_embedding.append(word)

print(f'Number of words with embeddings found: {len(with_embedding)}')
print(f'Number of words with embeddings NOT found: {len(without_embedding)}')
```

```python
# prepare input data for the model
# convert the train sentences to sequences of ids
train_input = vectorizer(train_text)
```

```python
# check the dimensions
# looks good: number examples and 100 max lenght
train_input.shape
```

```python
# do the same for development and test data
dev_input = vectorizer(dev_text)
test_input = vectorizer(test_text)
```

## Model Task A

```python
# delete model if exists
try:
    del CNN_model_A
except:
    pass
```

```python
# define the model

# define input layer
inputs = tf.keras.layers.Input(shape=(max_sequence_lenght,), name='inputs')

# define embedding layer
embedding_layer = tf.keras.layers.Embedding(embedding_matrix.shape[0],
                                            embedding_matrix.shape[1],
                                            embeddings_initializer=tf.keras.initializers.Constant(embedding_matrix),
                                            input_length=max_sequence_lenght,
                                            trainable=False)

# embeddings layer acting on inputs
embeddings = embedding_layer(inputs)

# Filter region size = 2
conv_layer_2 = tf.keras.layers.Conv1D(filters=10, 
                                      kernel_size=2, 
                                      strides=1, 
                                      padding='same',
                                      activation='relu', 
                                      use_bias=True,
                                      kernel_initializer='glorot_uniform', 
                                      bias_initializer='zeros',
                                      name='conv_2')

# convolution size 2 acting on embeddings
conv_out_2 = conv_layer_2(embeddings)

# Filter region size = 3
conv_layer_3 = tf.keras.layers.Conv1D(filters=10, 
                                      kernel_size=3, 
                                      strides=1, 
                                      padding='same',
                                      activation='relu', 
                                      use_bias=True,
                                      kernel_initializer='glorot_uniform', 
                                      bias_initializer='zeros',
                                      name='conv_3')

# convolution size 3 acting on embeddings
conv_out_3 = conv_layer_3(embeddings)

# Filter region size = 4
conv_layer_4 = tf.keras.layers.Conv1D(filters=10, 
                                      kernel_size=4, 
                                      strides=1, 
                                      padding='same',
                                      activation='relu', 
                                      use_bias=True,
                                      kernel_initializer='glorot_uniform', 
                                      bias_initializer='zeros',
                                      name='conv_4')

# convolution size 4 acting on embeddings
conv_out_4 = conv_layer_4(embeddings)

# pooling layers- maximum 
max_conv_2 = tf.keras.layers.GlobalMaxPooling1D(name='max_2')(conv_out_2)
max_conv_3 = tf.keras.layers.GlobalMaxPooling1D(name='max_3')(conv_out_3)
max_conv_4 = tf.keras.layers.GlobalMaxPooling1D(name='max_4')(conv_out_4)

# concatenate results from pooling
concat = tf.keras.layers.Concatenate(name='concat', axis=1)([max_conv_2, max_conv_3, max_conv_4])

# dropout layer
dropout = tf.keras.layers.Dropout(rate=0.2)

# dropout layer acting on concatenation
dropout_out = dropout(concat)

# dense layer
dense = tf.keras.layers.Dense(100, activation='relu', name='dense_layer')

# dense layer acting on dropout result
dense_out = dense(dropout_out)

# another dropout layer
dropout2 = tf.keras.layers.Dropout(rate=0.2)

# dropout layer acting on dense layer result
dropout2_out = dropout2(dense_out)

# output layer- softmax of 3 classes
softmax = tf.keras.layers.Dense(3, activation='softmax', name='softmax_layer')

# softmax layer acting on the output of previous dropout
softmax_out = softmax(dropout2_out)

# finally define the model with input and output
CNN_model_A = tf.keras.models.Model(inputs=inputs, outputs=softmax_out)
```

```python
# summary of the model
CNN_model_A.summary()
```

```python
# compile the model
# categorical_crossentropy: loss function appropiate for multi-class single label classification
# categorical accuracy: calculates how often predictions match one-hot labels
CNN_model_A.compile(loss='categorical_crossentropy', optimizer='adam', 
                    metrics=[tf.keras.metrics.categorical_accuracy, 'categorical_crossentropy'])
```

```python
# train the model and store the training history
training_history = CNN_model_A.fit(train_input, 
                                   train_labels_a_enc, 
                                   validation_data=(dev_input, dev_labels_a_enc),
                                   epochs=20, 
                                   verbose=0)
```

```python
# plot loss and accuracy
loss_accuracy_plots(training_history, xrange=20, task='A')
```

```python
# Evaluate model on DEVELOPMENT data

# predict labels using the development data
pred_labels_dev = from_prob_to_labels(CNN_model_A, dev_input, 'A')

# Create confusion matrix
confusion_matrix_plot(CNN_model_A, dev_input, dev_labels_a, task='A', normalize='true')

# Print the precision, recall, f1
print(metrics.classification_report(dev_labels_a, pred_labels_dev, digits=3))

# print global weighted f-1 score
print(f"Weighted f1-score: {metrics.f1_score(dev_labels_a, pred_labels_dev, pos_label=1, average='weighted')}")
```

```python
# first we need to define a hypermodel using a model builder function
# same model architecture we used earlier

def model_builder(hp):
    
    # hyperparameters
    # dropout after concatenation
    dropout_1 = hp.Float('dropout1', min_value=0.2, max_value=0.8, default=0.0, step=0.1)
    # droput after dense layer
    dropout_2 = hp.Float('dropout2', min_value=0.2, max_value=0.4, default=0.0, step=0.1)
    # number of filters of size 2
    filters2 = hp.Int('filters2', min_value=10, max_value=200, step=20)
    # number of filters of size 3
    filters3 = hp.Int('filters3', min_value=10, max_value=200, step=20)
    # number of filters of size 4
    filters4 = hp.Int('filters4', min_value=10, max_value=200, step=20)
    # number of neurons in dense layer
    dense_units = hp.Int('dense_units', min_value=10, max_value=150, step=10)
    
    # define input layer
    inputs = tf.keras.layers.Input(shape=(max_sequence_lenght,), name='inputs')

    # define embedding layer
    embedding_layer = tf.keras.layers.Embedding(embedding_matrix.shape[0],
                                                embedding_matrix.shape[1],
                                                embeddings_initializer=tf.keras.initializers.Constant(embedding_matrix),
                                                input_length=max_sequence_lenght,
                                                trainable=False)

    # embeddings layer acting on inputs
    embeddings = embedding_layer(inputs)

    # Filter region size = 2
    conv_layer_2 = tf.keras.layers.Conv1D(filters=filters2, 
                                          kernel_size=2, 
                                          strides=1, 
                                          padding='same',
                                          activation='relu', 
                                          use_bias=True,
                                          kernel_initializer='glorot_uniform', 
                                          bias_initializer='zeros',
                                          name='conv_2')

    # convolution size 2 acting on embeddings
    conv_out_2 = conv_layer_2(embeddings)

    # Filter region size = 3
    conv_layer_3 = tf.keras.layers.Conv1D(filters=filters3, 
                                          kernel_size=3, 
                                          strides=1, 
                                          padding='same',
                                          activation='relu', 
                                          use_bias=True,
                                          kernel_initializer='glorot_uniform', 
                                          bias_initializer='zeros',
                                          name='conv_3')

    # convolution size 3 acting on embeddings
    conv_out_3 = conv_layer_3(embeddings)

    # Filter region size = 4
    conv_layer_4 = tf.keras.layers.Conv1D(filters=filters4, 
                                          kernel_size=4, 
                                          strides=1, 
                                          padding='same',
                                          activation='relu', 
                                          use_bias=True,
                                          kernel_initializer='glorot_uniform', 
                                          bias_initializer='zeros',
                                          name='conv_4')

    # convolution size 4 acting on embeddings
    conv_out_4 = conv_layer_4(embeddings)

    # pooling layers- maximum 
    max_conv_2 = tf.keras.layers.GlobalMaxPooling1D(name='max_2')(conv_out_2)
    max_conv_3 = tf.keras.layers.GlobalMaxPooling1D(name='max_3')(conv_out_3)
    max_conv_4 = tf.keras.layers.GlobalMaxPooling1D(name='max_4')(conv_out_4)

    # concatenate results from pooling
    concat = tf.keras.layers.Concatenate(name='concat', axis=1)([max_conv_2, max_conv_3, max_conv_4])

    # dropout layer
    dropout = tf.keras.layers.Dropout(rate=dropout_1)

    # dropout layer acting on concatenation
    dropout_out = dropout(concat)

    # dense layer
    dense = tf.keras.layers.Dense(dense_units, activation='relu', name='dense_layer')

    # dense layer acting on dropout result
    dense_out = dense(dropout_out)

    # another dropout layer
    dropout2 = tf.keras.layers.Dropout(rate=dropout_2)

    # dropout layer acting on dense layer result
    dropout2_out = dropout2(dense_out)

    # output layer- softmax of 3 classes
    softmax = tf.keras.layers.Dense(3, activation='softmax', name='softmax_layer')

    # softmax layer acting on the output of previous dropout
    softmax_out = softmax(dropout2_out)

    # finally define the model with input and output
    model = tf.keras.models.Model(inputs=inputs, outputs=softmax_out)

    # compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', 
                  metrics=[tf.keras.metrics.categorical_accuracy, 'categorical_crossentropy'])
    
    return model
```

```python
# define the model - increased dropout rates

# define input layer
inputs = tf.keras.layers.Input(shape=(max_sequence_lenght,), name='inputs')

# define embedding layer
embedding_layer = tf.keras.layers.Embedding(embedding_matrix.shape[0],
                                            embedding_matrix.shape[1],
                                            embeddings_initializer=tf.keras.initializers.Constant(embedding_matrix),
                                            input_length=max_sequence_lenght,
                                            trainable=False)

# embeddings layer acting on inputs
embeddings = embedding_layer(inputs)

# Filter region size = 2
conv_layer_2 = tf.keras.layers.Conv1D(filters=150, 
                                      kernel_size=2, 
                                      strides=1, 
                                      padding='same',
                                      activation='relu', 
                                      use_bias=True,
                                      kernel_initializer='glorot_uniform', 
                                      bias_initializer='zeros',
                                      name='conv_2')

# convolution size 2 acting on embeddings
conv_out_2 = conv_layer_2(embeddings)

# Filter region size = 3
conv_layer_3 = tf.keras.layers.Conv1D(filters=130, 
                                      kernel_size=3, 
                                      strides=1, 
                                      padding='same',
                                      activation='relu', 
                                      use_bias=True,
                                      kernel_initializer='glorot_uniform', 
                                      bias_initializer='zeros',
                                      name='conv_3')

# convolution size 3 acting on embeddings
conv_out_3 = conv_layer_3(embeddings)

# Filter region size = 4
conv_layer_4 = tf.keras.layers.Conv1D(filters=170, 
                                      kernel_size=4, 
                                      strides=1, 
                                      padding='same',
                                      activation='relu', 
                                      use_bias=True,
                                      kernel_initializer='glorot_uniform', 
                                      bias_initializer='zeros',
                                      name='conv_4')

# convolution size 4 acting on embeddings
conv_out_4 = conv_layer_4(embeddings)

# pooling layers- maximum 
max_conv_2 = tf.keras.layers.GlobalMaxPooling1D(name='max_2')(conv_out_2)
max_conv_3 = tf.keras.layers.GlobalMaxPooling1D(name='max_3')(conv_out_3)
max_conv_4 = tf.keras.layers.GlobalMaxPooling1D(name='max_4')(conv_out_4)

# concatenate results from pooling
concat = tf.keras.layers.Concatenate(name='concat', axis=1)([max_conv_2, max_conv_3, max_conv_4])

# dropout layer
dropout = tf.keras.layers.Dropout(rate=0.75)

# dropout layer acting on concatenation
dropout_out = dropout(concat)

# dense layer
dense = tf.keras.layers.Dense(110, activation='relu', name='dense_layer')

# dense layer acting on dropout result
dense_out = dense(dropout_out)

# another dropout layer
dropout2 = tf.keras.layers.Dropout(rate=0.7)

# dropout layer acting on dense layer result
dropout2_out = dropout2(dense_out)

# output layer- sigmoid for binary classification
sigmoid = tf.keras.layers.Dense(1, activation='sigmoid', name='sigmoid_layer')

# sigmoid layer acting on the output of previous dropout
sigmoid_out = sigmoid(dropout2_out)

# finally define the model with input and output
CNN_model_B_optimized = tf.keras.models.Model(inputs=inputs, outputs=sigmoid_out)
```

```python
# compile the model
# binary_crossentropy: loss function appropiate for binary classification
# binary accuracy: calculates how often predictions match binary labels
CNN_model_B_optimized.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy', 'binary_crossentropy'])
```
